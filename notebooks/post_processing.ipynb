# QC. make jpg from dicom for retinal photography folder for visual inspection QC by each site

```python
import os
import sys
from tqdm import tqdm

sys.path.append("/Users/nayoonkim/pipeline_imaging/aireadi_retinal_imaging/year_3")
import organize_utils

import numpy as np
from joblib import Parallel, delayed
from tqdm import tqdm
import numpy as np
from joblib import Parallel, delayed
from tqdm import tqdm
from multiprocessing import Manager
import os


def split_into_n_parts(lst, n):
    """Split list into n nearly equal parts"""
    return np.array_split(lst, n)


def process_sublist(sublist, sublist_index, dst_root):
    ok, fail = 0, 0
    for f in sublist:
        try:
            organize_utils.file_to_jpg(f, dst_root)
            ok += 1
        except Exception as e:
            fail += 1
    return ok, fail


def dicom_to_jpg(files, output_folder):
    print("Total files:", len(files))

    sublists = split_into_n_parts(files, 100)

    results = Parallel(n_jobs=10, backend="loky")(
        delayed(process_sublist)(sublist, i, output_folder)
        for i, sublist in enumerate(sublists)
    )

    total_ok = sum(ok for ok, _ in results)
    total_fail = sum(fail for _, fail in results)
    print(f"Done. Successful: {total_ok}  Failed: {total_fail}")


def process_dicom_to_jpb_per_site(device, modality, input, output_root):

    dicom_folder = f"{input}/{modality}/{device}"

    files = organize_utils.get_dcm_files(dicom_folder)

    list_1 = []
    list_4 = []
    list_7 = []
    others = []

    for f in files:
        base = os.path.basename(f)
        if base.startswith("1"):
            list_1.append(f)
        elif base.startswith("4"):
            list_4.append(f)
        elif base.startswith("7"):
            list_7.append(f)
        else:
            others.append(f)

    print("Starts with 1:", len(list_1))
    print("Starts with 4:", len(list_4))
    print("Starts with 7:", len(list_7))

    if others:
        print("\nFiles with unexpected IDs:")
        for o in others:
            print(o)
    else:
        print("\n✅ All files start with 1, 4, or 7.")

    uw_folder = os.path.join(output_root, "UW", device, modality)
    ucsd_folder = os.path.join(output_root, "UCSD", device, modality)
    uab_folder = os.path.join(output_root, "UAB", device, modality)

    dicom_to_jpg(list_1, uw_folder)
    dicom_to_jpg(list_4, ucsd_folder)
    dicom_to_jpg(list_7, uab_folder)
```

# pixel QC 

```python
import os
import pydicom
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed


def get_filtered_all_file_names(folder_path):
    """
    Retrieves a list of all .dcm file paths in a directory, excluding '._' prefixed files.
    """
    filtered_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if not file.startswith("._") and file.endswith(".dcm"):
                filtered_files.append(os.path.join(root, file))
    return filtered_files


def _check_one_file(f):
    """
    Helper to check one DICOM file. Returns error message if failed, None otherwise.
    """
    try:
        dataset = pydicom.dcmread(f)
        _ = dataset.pixel_array  # force pixel data load
        return None
    except Exception as e:
        return f"Error reading file {f}: {e}"


def check_pixel_array(folder, max_workers=8):
    """
    Walk through all .dcm files in the folder, attempt to load pixel arrays in parallel.
    Only prints files that fail.
    """
    filelist = get_filtered_all_file_names(folder)

    errors = []
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = {executor.submit(_check_one_file, f): f for f in filelist}
        for future in tqdm(
            as_completed(futures), total=len(futures), desc="Checking DICOMs"
        ):
            result = future.result()
            if result:
                errors.append(result)

    if not errors:
        print("All files read successfully.")
    else:
        print("Some files failed:")
        for err in errors:
            print(err)


# Example usage
folder = "/Volumes/2024release/sample_data/imaging"
check_pixel_array(folder, max_workers=12)
```
